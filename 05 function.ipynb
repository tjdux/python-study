## Function

```python
def my_func(country):
    print(f"Helllo world, {country}")

my_func(country="USA") # country: parameter, USA: argument
my_func("USA")
```

## parameters

```python
def travel_to_country(country: str, name: str):
    print(f"Hello {name}")
    print(f"You're going to travel to {country}")
    print("Are you excited?")

# positional argument
travel_to_country("USA", "Jane")

# keyword argument
travel_to_country(country="France", name="Jack")
travel_to_country(name="Doe", country="Germany")
```

## return

```python
def get_name(first_name, last_name):
    if first_name == "":
        return "Your first name is missing ..."
    if last_name == "":
        return "Your last name is missing ..."
    return f"{first_name}, {last_name}"

return_value = get_name(
    first_name=input("Your first name?"),
    last_name=input("Your last name?"))

print(return_value)
```

## Mini Proj: Going Dutch

```python
num_of_ppl = 3
food_prices = [10, 20, 15, 17, 50]

def going_dutch(food_prices: list, num_of_ppl: int):
    return sum(food_prices)/num_of_ppl

print(going_dutch(food_prices=food_prices, num_of_ppl=num_of_ppl))
```

## Mini Proj: Prime number

```python
import math

is_this_prime_number = 4

def is_prime_number(num: int):
    if num <= 1:
        return False

    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True
    
print(is_prime_number(num=is_this_prime_number))
print(is_prime_number(num=7919))
```

## Higher Order Function

- A function is an instance of the Object type.
- You can store the function in a variable.
- You can pass the function as a parameter to another function.
- You can return the function from a function.
- You can store them in data structures such as hash tables,

```python
def add(num1, num2):
    return num1 + num2

add2 = add

def subtract(num1, num2):
    return num1 - num2


def multiply(num1, num2):
    return num1 * num2


def divide(num1, num2, is_floor=True):
    if is_floor:
        return num1 // num2
    else:
        return num1 / num2


def calc(num1, num2, func):
    return func(num1, num2)

print(calc(num1=10, num2=5, func=add2))
```

```python
# Decorator
# Wrap another functions to extend  the functionality without change the original code

def higher_order_example(func):
    def inside():
        print("start")
        func()
        print("end")
    return inside

@higher_order_example
def sample_example():
    print("I am inside")

sample_example()
```

## Advanced Python arguments

```python
# Arguments with Default Values

def write_str_to_file(file, content, mode="w"):
    with open(file, mode=mode) as f:
        f.write(content)
        
```

```python
# Unlimited Positional Arguments

def unlimited_func(*args):
    print(args)

unlimited_func(1, 2, 3)
unlimited_func(1, 2, 3, 4, 5)

def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3))
```

```python
# **kwargs: Keyworded arguments
def calc(**kwargs):
    print(kwargs)
    print(kwargs["n1"])

calc(n1=1, n2=2, func="add")
```

```python
def calc(**kwargs):
    for k, v in kwargs.items():
        print(f"{k}: {v}")

    if kwargs['func'] == "add":
        return kwargs['n1'] + kwargs['n2']

print(calc(n1=1, n2=3, func="add"))
```

```python
# func: mandatory argument
def calc(func, **kwargs):
    if func == "add":
        return kwargs['n1'] + kwargs['n2']

print(calc(n1=1, n2=3, func="add"))
```
