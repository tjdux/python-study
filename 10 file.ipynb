## File Management

```python
file = open("sample.md", "r", encoding="utf-8")
print(file.read())
file.close() # GC; Garbage Collector
```

```python
# no close anymore
with open("sample.md", "r", encoding="utf-8") as file:
    print(file.read())
```

```python
# "w": overwrite / "a": append
with open("WRITEME.txt", "w") as file:
    print(file.write("Thanks"))
```

```python
import csv

with open("sample.csv", "r") as data_file:
    sample_data = csv.reader(data_file)
    for row in sample_data:
        if row[0] != "country":
            print(row[0])
```

## Pandas

```python
import pandas as pd

data = pd.read_csv("sample.csv")

print(data)
```

```python
# DataFrame: General 2D labeled, size-mutable tabular structure with potentially heterogeneously-typed column
print(type(data))
print(data.to_dict())
```

```python
# Series: 1D labeld homogeneously-typed array
print(type(data["country"]))
print(data["country"].to_list())
```

```python
# different way to access to the column
print(data.population)
```

```python
print(f"sum of population: {sum(data["population"].to_list())}")
print(f"max of population: {data["population"].max()}")
```

```python
# row data
print(data[data.country == "USA"])
print(data[data.population == data.population.max()])
```
