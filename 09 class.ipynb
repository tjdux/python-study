## Class constructor

```python
class Car:
    def __init__(self, color, brand):
        self.color = color
        self.brand = brand
        

tesla = Car("red", "tesla")
tesla.engine_type = "electric"

print(vars(tesla))
```

## Method

```python
class Car:
    def __init__(self, color, engine_type):
        self.color = color
        self.engine_type = engine_type
        self.speed = 0
        self.is_start = False
    
    def start_engine(self):
        self.speed = 0
        self.is_start = True

    def speed_up(self, speed):
        self.speed += speed

    def speed_down(self, speed):
        self.speed -= speed


car = Car(color="green", engine_type="electric")
print(vars(car))
car.start_engine()
print(vars(car))
car.speed_up(150)
print(vars(car))
car.speed_down(50)
print(vars(car))
```

## Inheritance

```python
class Car:
    def __init__(self):
        self.wheel_count = 4
        self.door_count = 2

    def start(self):
        print("started...")

    def drive(self):
        print("driving...")


class ElectricCar(Car):
    def __init__(self):
        super().__init__()

    def start(self):
        super().start()
        print("No sound ...")


class CombustionEngineCar(Car):
    def __init__(self):
        super().__init__()

    def start(self):
        super().start()
        print("vroooom...")


ec = ElectricCar()
ec.start()
ec.drive()

print("-----------")

cec1 = CombustionEngineCar()
cec1.start()
cec1.drive()
print(id(cec1))

cec2 = CombustionEngineCar()
print(id(cec2))
```
