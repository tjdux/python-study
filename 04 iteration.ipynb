## FizzBuzz

```python
def fizz_buzz(n):
    answer = []
    for i in range(1, n+1):
        if i % 15 == 0:
            answer.append("FizzBuzz")
        elif i % 3 == 0:
            answer.append("Fizz")
        elif i % 5 == 0:
            answer.append("Buzz")
        else:
            answer.append(f"{i}")
    
    return answer

print(fizz_buzz(3))
print(fizz_buzz(5))
print(fizz_buzz(15))
```

## Mini Proj: Strong Password

```python
lower_alphabets = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
upper_alphabets = []

for alphabet in lower_alphabets:
    upper_alphabets.append(alphabet.upper())

alphabets = lower_alphabets + upper_alphabets
numbers = [str(i) for i in range(0, 10)]
symbols = ['!', '#', '$', '%', '&', '(', ')', '*', '+']
```

```python
import random

num_of_alphabets = 9
num_of_numbers = 3
num_of_symbols = 2

strong_password = []

items = [num_of_alphabets, num_of_numbers, num_of_symbols]

for i in range(0, num_of_alphabets):
    strong_password.append(alphabets[random.randint(0, len(alphabets)-1)])

for i in range(0, num_of_numbers):
    strong_password.append(numbers[random.randint(0, len(numbers)-1)])

for i in range(0, num_of_symbols):
    strong_password.append(symbols[random.randint(0, len(symbols)-1)])

random.shuffle(strong_password)

print("".join(strong_password))
```

## Mini Proj: Hangman

```python
question = list("batman")
lst = []

for i in range(0, len(question)):
    lst.append("_")

print(f"Here is the word {lst}")

num_of_try = 10

while True:
    char = input("Guess the char?")
    for i in range(0, len(question)):
        if char == question[i]:
            lst[i] = char

    if "_" not in lst:
        print("You won!")
        break
    
    if num_of_try < 1:
        print("You lost!")
        break
    
    print(lst)
    num_of_try -= 1
```

## Mini Proj: Reverse String

```python
value = "Hello World"

# 1️⃣ reverse
value_list = list(value)
value_list.reverse()
print("".join(value_list))
```

```python
# 2️⃣ reversed
print("".join(list(reversed(value))))
```

```python
# 3️⃣ forLoop
val = ""
for i in range(len(value)-1, -1, -1):
    val += value[i]
print(val)
```

```python
# 4️⃣ while
value_list = list(value)
val = ""
while len(value_list) > 0:
    val += value_list.pop()
print(val)
```
