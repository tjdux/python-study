## List
- Grouping
- Keep the order
- mutable

```python
countries = ["South Korea", "USA", "Japan", "China"]

# ìš”ì†Œ ë³€ê²½
countries[2] = "Vietnam"
print(countries)

print(f"countries[len(countries)-1]: {countries[len(countries)-1]}")
print(f"countries[-2]: {countries[-2]}")

# pop - ì›ë³¸ ë°°ì—´ ë³€ê²½
print(countries.pop())
print(countries)

print(countries.pop(0))
print(countries)
```

```python
element = "c" 
alphabets = ["b", element, "d"]
print(alphabets)

alphabets.append("e")
print(alphabets)

alphabets += ["f", "g"]
print(alphabets)

alphabets.insert(0, "a")
print(alphabets)
```

```python
# Nested List
alphabets = [["a", "b"],["c"]]
print(alphabets)

for i in range(len(alphabets)):
    for j in range(len(alphabets[i])):
        print(f"alphabets[{i}][{j}]: {alphabets[i][j]}")
```

## Dictionary
- mutable
- ordered

```python
dic = {
    "country": "south korea",
    "city": "seoul",
    "gender": "female",
    "age": 28
}
print(dic["country"])
```

## Mini Proj: Grading Program

```python
students = {
    "tom": 82,
    "jerry": 93,
    "joon": 99,
    "misha": 80,
    "amy": 74
}

def get_grade(score: int):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "F" 
    
res = {}
for student, score in students.items():
    res[student] = get_grade(score)

print(res)
```

## Nested Data Structure

```python
usa = {
    "name": "USA",
    "continent": "North America"
}

south_korea = {
    "name": "South Korea",
    "continent": "East Asia"
}

countries = [
    usa,
    south_korea
]

print(countries)
```

## Collections module

### Counter

```python
from collections import Counter

lst = [1,2,3,2,2,2,3,3,3,1,1,1,1,2,2,2,3]
c = Counter(lst)
print(c)

print(dict(c))
print(list(c))
```

```python
c = Counter("hello")
print(c)
```

```python
sentences = "hello world what a wonderful world"
c = Counter(sentences.split(" "))
print(c)

# most common
print(c.most_common)
```

### defaultdict

```python
# defaultdict never raises a KeyError
from collections import defaultdict

d = defaultdict(lambda: 0) # default ê°’
d["here"] = 10

print(d)

d["there"]
print(d["there"])
```

### namedtuple

```python
from collections import namedtuple

Dog = namedtuple("Dog", ["age", "breed", "name"])

sammy = Dog(age=5, breed="Husky", name="Sammy")
print(sammy)
print(sammy.age)
print(sammy[0])
```
