## List
- Grouping
- Keep the order

```python
countries = ["South Korea", "USA", "Japan", "China"]

# 요소 변경
countries[2] = "Vietnam"
print(countries)

print(f"countries[len(countries)-1]: {countries[len(countries)-1]}")
print(f"countries[-2]: {countries[-2]}")

# pop - 원본 배열 변경
print(countries.pop())
print(countries)

print(countries.pop(0))
print(countries)
```

```python
element = "c" 
alphabets = ["b", element, "d"]
print(alphabets)

alphabets.append("e")
print(alphabets)

alphabets += ["f", "g"]
print(alphabets)

alphabets.insert(0, "a")
print(alphabets)
```

```python
# Nested List
alphabets = [["a", "b"],["c"]]
print(alphabets)

for i in range(len(alphabets)):
    for j in range(len(alphabets[i])):
        print(f"alphabets[{i}][{j}]: {alphabets[i][j]}")
```

## Dictionary

```python
dic = {
    "country": "south korea",
    "city": "seoul",
    "gender": "female",
    "age": 28
}
print(dic["country"])
```

```python
print("name" not in dic)
print("age" in dic)
```

```python
for key, value in dic.items():
    print(f"{key}: {value}")
```

```python
for key in dic:
    print(f"{key}: {dic[key]}")
```

```python
dic1 = {
    1: 1,
    2: 3,
}
print(dic1[2])
print(2 in dic1)
```

## Mini Proj: Grading Program

```python
students = {
    "tom": 82,
    "jerry": 93,
    "joon": 99,
    "misha": 80,
    "amy": 74
}

def get_grade(score: int):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "F" 
    
res = {}
for student, score in students.items():
    res[student] = get_grade(score)

print(res)
```

## Nested Data Structure

```python
usa = {
    "name": "USA",
    "continent": "North America"
}

south_korea = {
    "name": "South Korea",
    "continent": "East Asia"
}

countries = [
    usa,
    south_korea
]

print(countries)
```

```python
for country in countries:
    print(country)
    print(country["name"])
```

```python
california = ["los angeles", "san francisco"]
us_states = [california, "washington", "oregon"]
print(us_states)
```

```python
countries = [
    us_states,
    south_korea
]
print(countries)
```
