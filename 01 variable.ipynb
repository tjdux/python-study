## Primitive data type

```python
# Integers
profile_number = 2030

# floats - decimal point
score = float(59.5)

# boolean
is_correct = True

# string
teacher_name = "Cisca"
course_name = 'CSC'
lecture_name = """
Cisca
Oladipo
"""
print(lecture_name)

# string subscript
print(teacher_name[len(teacher_name)-1])

# check object type
print(type(profile_number))
print(type(score))
print(type(is_correct))
print(type(teacher_name))
```

## Mathmatical operation

```python
print(f"3 + 1 = {3 + 1}")
print(f"3 - 1 = {3 - 1}")
print(f"3 * 1 = {3 * 1}")

# ✨ int / int여도 항상 결과는 float!
print(f"3 / 2 = {3 / 2}")
print(f"3 / 1 = {3 / 1}") 

# floor division(소수점을 없앰)
print(f"7 // 2 = {7 // 2}")

# 지수(Exponents)
print(f"3**3 = {3**3}")

# 나머지 값
print(f"60 % 13 = {60 % 13}")
print(f"12.5 % 5.5 = {12.5 % 5.5}")

# # PEMDAS
# Parentheses
# Exponents
# Multiplication
# Division
# Addition
# Subtraction
# And Left to Right

sum = 10
sum += 1 # sum = sum + 1
sum += 2
print(sum)
```

## f-string

```python
name = "Jane"
age = 30

# Python 2.6
print("Hello, {}. I am {}.".format(name, age))
print("Hello, {1}. I am {0}.".format(age, name))

person = {'name': 'Jane', 'age': 20}
print("Hello, {name}. I am {age}.".format(name=person['name'], age=person['age']))
print("Hello, {name}. I am {age}.".format(**person))
# **: dictionary unpacking
# *: list, tuple unpacking

# Python 3.6
print(f"Hello, {name}. I am {age}.")
print(F"Hello, {name}. I am {age}.")
print(f"{name.lower()} is cool.")
```
