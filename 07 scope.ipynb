```python
my_score = 50 # global scope

def inside_value_function():
    # global 키워드를 사용해 global scope 접근 가능
    my_score = 80 # local scope
    print(f"my score inside is {my_score}")

inside_value_function()
print(f"my score inside is {my_score}")
```

```python
did_extra_work = True

# if, for: 새로운 namespace를 생성하지 않음
if did_extra_work:
    my_score = 90

print(f"my score outside is {my_score}")
```

```python
# nonlocal: enclosed function이 enclosing function의 변수에 접근할 때
def a():
    x = 10
    def b():
        nonlocal x
        x = 20
    b()
    print(x)

a()
```

```python
country = ["south korea"]

def func1():
    country.append("usa")

func1()
print(country)

def func2():
    country = ["usa"]

func2()
print(country)
```
