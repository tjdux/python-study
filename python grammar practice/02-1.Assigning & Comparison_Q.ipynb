## ğŸ§  Understanding `id()` and Variable References in Python
### 1. The Key Point

In Python, variables donâ€™t have IDs â€” objects do.
The `id()` function returns the unique identifier of the object a variable refers to, not of the variable itself.

### 2. Step-by-Step Example
```python

y = x
```


- Python creates one list object in memory (e.g., at address `0x1234`).

- The name `x` is bound to that list.

- When you do `y = x`, youâ€™re not copying the list â€”
youâ€™re just creating another name (`y`) that points to the same list object.

|Name|	Refers to Object|	Objectâ€™s id()|
|---|---|---|
|`x`|	â†’ list at `0x1234`|	`0x1234`|
|`y`|	â†’ same list at `0x1234`|	`0x1234`|

So:
```python
id(x) == id(y)   # True
x is y           # True
```

### 3. Visual Analogy

Think of variables as labels attached to boxes (objects):

```css
x  â”€â”
    â”œâ”€â”€>  ["orange", "banana", "apple"]

```

Both `x` and `y` are names pointing to the same box.
When you call `id(x)` or `id(y)`, youâ€™re asking for the identity of that box, not the labels.

### 4. Why â€œVariable IDsâ€ Donâ€™t Exist

In lower-level languages (like C or Java), a variable corresponds to a memory slot.
Python hides that â€” it treats variables as names in a namespace (internally, a dictionary mapping names to objects).

You can confirm this:
```python

y = x
print(id(x), id(y))   # same

x = ["b"]             # rebinds x to a new list
print(id(x), id(y))   # different now
```


When you reassign `x`, you donâ€™t change the object â€”
you just change what name `x` points to.

### 5. âœ… Summary
|Concept|	Description|

**`id()` measures object identity, not variable identity.**
