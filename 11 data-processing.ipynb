## List comprehension

```python
country_list = ["USA", "South Korea", "Japan"]

lower_case_list = [country.lower() for country in country_list]

print(lower_case_list)
```

```python
# string list comprehension

sample = "silicon valley"
print([ch for ch in sample])
```

```python
# conditional list comprehension

sampling = [2, 3, 1, 1, 2, 3, 4]
filtered_sampling = [sample for sample in sampling if sample > 1]
print(filtered_sampling)
```

## Dictionary comprehension

```python
incorrect_score_dict = {
    "Tom": 80,
    "Lisa": 75,
    "Sarah": 90
}

correct_score_dict = {name: score+5 for (name, score) in
                        incorrect_score_dict.items()}
print(correct_score_dict)
```

## Iterate Pandas dataframe

```python
score_dict = {
    "student": ["Tom", "Lisa", "Sarah"],
    "score": [80, 90, 95]
}

[print(col) for (col, _) in score_dict.items()]
```

```python
import pandas as pd

score_df = pd.DataFrame(score_dict)

# (key, value)
for (student, score) in score_df.items():
    print(student)
    print(score)
```

```python
# loop through rows

for (i, row) in score_df.iterrows():
    print(f"{row.student}: {row.score}")
```

## Dynamic & Strong Typing
- Dynamic typing means that runtime objects (values) have a type, as opposed to static typing where variables have a type.
- Strong typing means that variables do have a type and that the type matters when performing operations on a variable.
